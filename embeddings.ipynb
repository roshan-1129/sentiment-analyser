```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Example usage
sent=['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am going to the market',
        'i am agood boy',
        'i am agood developer',
        'understand the meaning of words']
```

```python
#define vocabulary size
voca_size=10000
```

```python
#one hot representation 
onehot_repr=[one_hot(words,voca_size)for words in sent]
for item in onehot_repr:
    print(item)
```

```python
#word embeddings representation can be done using Embedding layer in keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
##feature represntation 
dim=10
```

```python
model=Sequential()
model.add(Embedding(voca_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
